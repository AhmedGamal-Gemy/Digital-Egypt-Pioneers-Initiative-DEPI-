```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

data = pd.read_csv('/content/titanic_data.csv')
data.head(20)
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())

data = data.drop('Cabin',axis=1)
data = data.drop('Embarked',axis=1)

data.isna().sum()
```

```python
data = data.drop_duplicates()
```

```python
sns.swarmplot(x=data.Survived,y= data.Age)
```

```python
sns.pointplot(x=data.Survived,y= data.Age)
```

```python
sns.barplot(x=data.Survived,y= data.Age)
```

```python
sns.boxplot(x=data.Survived,y= data.Age)
```

```python
numeric_data = data.select_dtypes(include=np.number)
sns.heatmap(numeric_data.corr(),cmap='Blues')
```

```python
sns.histplot(data.Age,kde=True)
```

```python
sns.jointplot(x=data.Age,y=data.Survived,kind='scatter')
```
